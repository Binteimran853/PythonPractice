```
# Time Complexity= O(n^2) Space = O(n)
def return_kth_sorted(num, k):
    lst = []
    while num:
        i = 0
        digit = num % 10
        while i < len(lst) and lst[i] < digit:
            i += 1
        lst.insert(i, digit)
        num //= 10
    return lst[:k]

print(return_kth_sorted(16423, 4))
```

```
# Time Complexity= O(n + klogn) Space = O(n)
import heapq
def return_kth_sorted(num, k):
   h = []
   while num:
       heapq.heappush(h, num % 10)
       num //= 10
   return heapq.nsmallest(k, h)

print(return_kth_sorted(16423, 3))
```

```
def return_kth_sorted(k, nums):
    if k <=0 or k > len(str(nums)):
        return -1
    sorted_num = 0
    count = [0] * 10    #because number range will be between 0-9
    while nums > 0:
        count[nums % 10] += 1
        nums //= 10
    for i in range(10):
        while count[i] > 0 :
            if k == 0:
                return sorted_num
            sorted_num = sorted_num * 10 + i
            count[i] -= 1
            k -= 1
    return sorted_num
print(f'kth-Sorted values: {return_kth_sorted(3,16432)}')
# time complexity = O(n)
# space complexity = O(1)
```
