```
# Time Complexity= O(n^2) Space = O(n)
def return_kth_sorted(num, k):
    lst = []
    while num:
        i = 0
        digit = num % 10
        while i < len(lst) and lst[i] < digit:
            i += 1
        lst.insert(i, digit)
        num //= 10
    return lst[:k]

print(return_kth_sorted(16423, 4))
```

```
# Time Complexity= O(n + klogn) Space = O(n)
import heapq
def return_kth_sorted(num, k):
   h = []
   while num:
       heapq.heappush(h, num % 10)
       num //= 10
   return heapq.nsmallest(k, h)

print(return_kth_sorted(16423, 3))
```

```
# Time: O(n log n) (dominated by sort)
# Space: O(n)
def return_kth_sorted(num, k):
    lst = []
    lst.extend(list(str(num)))
    lst = [int(digit) for digit in lst]
    lst.sort()
    return lst[:k]

print(return_kth_sorted(1642334542424, 4))
```
