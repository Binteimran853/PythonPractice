```python
# Work on List
class Grossery:
    def __init__(self, name):
        self.name = name
        self.items = []

    def add_item(self):
        name = input("Enter item name: ").split()
        self.items.append(name)

    def display_items(self):
        if not self.items:
            print("No items available")
        else:
            for item in self.items:
                print(f"Name: {item}")

    def delete_item(self):
        if not self.items:
            print("No items available")
        else:
            item_no = int(input("Enter item number to delete: "))
            self.items.remove(item_no)


grossery = Grossery('')

while True:
    print("=== Menu ===")
    print("1) Add item")
    print("2) View items")
    print("3) Delete items")
    print("4) Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        grossery.add_item()
    elif choice == 2:
        grossery.display_items()
    elif choice == 3:
        grossery.delete_item()
    elif choice == 4:
       break
    else:
        print("Invalid choice\ntry again")


```
