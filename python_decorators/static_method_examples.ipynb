```python
class Calculator:
    def __init__(self, name, version):
        self.name = name
        self.version = version

    def get_calculator_info(self):
        return f'{self.name} V.{self.version}'

    @staticmethod
    def add_numbers(*numbers):
        return sum(numbers)

calculator = Calculator(name = 'Calculator', version = '1.0')
print(calculator.get_calculator_info())

```

```python
lst = [('Ali', "uk"), ('Arslan', 'uk'), ('Ahmed', 'La')]

my_dict = dict()
for i in lst:
     name = i[0]
     city = i[1]
     if city not in my_dict:
         my_dict[city] = []
     my_dict[city].append(name)

print(my_dict)
```

```python
students= set()
students = {'Ali', 'Arslan', 'Ahmed', 'La'}
employee = students.copy()

print("student's set id:",id(students))
print("employee's set id:",id(employee))

print(students)
print(employee)

employee = students
employee.add('PC')
print("student's set id:",id(students))
print("employee's set id:",id(employee))

print(students)
print(employee)

employee = set(students)

print("student's set id:",id(students))
print("employee's set id:",id(employee))

employee.add('PC')

print(students)
print(employee)

x = (1, 2)
students = {x}
employee = students.copy()

print(id(students) == id(employee))      # False
print(id(x) == id(next(iter(employee)))) # True
```
