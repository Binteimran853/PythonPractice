# Stack Implementation

### using Array

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self,item):
        self.stack.append(item)

    def isEmpty(self):
        return len(self.stack) == 0

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.stack.pop()

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.stack[-1]

    def size(self):
        return len(self.stack)

myStack = Stack()
myStack.push('Ali')
myStack.push('Yaqoob')
myStack.push(231)
myStack.size()
print('Poped element:',myStack.pop())


```

### Using LinkList

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None


class Stack:
    def __init__(self):
        self.head = None
        self.size = 0

    def push(self, val):
        new_node = Node(val)
        if self.head:
            new_node.next = self.head
            self.head = new_node
        else:
            self.head = new_node
        self.size += 1

    def traverseAndPrint(self):
        if self.isEmpty():
            return "Stack is Empty"
        current = self.head
        while current:
            print(current.val,"->", end = '')
            current = current.next

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        current = self.head
        self.head = self.head.next
        self.size -= 1
        return current.val

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.head.val

    def stack_size(self):
        return self.size

    def isEmpty(self):
        return self.size == 0


myNewObj = Stack()
myNewObj.push('Atif')
myNewObj.push('Adnan')
myNewObj.push('Usama')
myNewObj.push('Ali')
myNewObj.push('Usman')
myNewObj.push('Yaqoob')
myNewObj.traverseAndPrint()
print('\nPop value: ',myNewObj.pop())
print('Top value: ',myNewObj.peek())
print('isEmpty: ',myNewObj.isEmpty())
print('Top value: ',myNewObj.stack_size())
```

## Queue Using LinkList

```python
class Node:
    def __int__(self, val):
        self.val = 0
        self.next = None


class Queue:
    def __init__(self):
        self.rear = None
        self.front = None
        self.size = 0

    def enqueue(self, val):
        newNode = Node(val)
        if self.rear == None:
            self.front = self.rear = newNode
        self.rear.next = newNode
        self.rear = newNode
        self.size += 1

    def isEmpty(self):
        return self.size == 0

    def dequeue(self):
        if self.isEmpty():
            return "Queue is Empty"
        temp = self.front
        self.front = self.front.next
        self.size -= 1
        return temp.val

    def peek(self):
        if self.isEmpty():
            return "Queue is Empty"
        return self.front.val

    def size(self):
        return self.size

    def traverse_and_print(self):
        if self.isEmpty():
            return "Queue is Empty"
        current = self.front
        while current:
            print(current.val,"->", end = '')
            current = current.next

myQueue = Queue()
myQueue.enqueue('Ali')
myQueue.enqueue('Yaqoob')
myQueue.enqueue(231)
myQueue.enqueue(34)
myQueue.enqueue(45)
myQueue.enqueue(56)
print(f'Poped values: {myQueue.dequeue()}->{myQueue.dequeue()}')
print(f'Peek Value: {myQueue.peek()}')
print("")
```

```python
def factorial (num):   #TIme and space Complexity: O(n) O(1)
    fact = 1
    if num == 0 or num == 1:
        return 1
    while num >= 1:
        fact *=  num
        num -= 1
    return fact
num = int(input("Enter a number to find factorical:"))
print(factorial(num))
```

```python
#HashTable implementation
hashTable = ['None','None','None','None','None','None','None','None','None','None',]
def hash(value):
    sum_of_char = 0
    for char in value:
        sum_of_char += ord(char)
    index = sum_of_char % 10
    return index

def add(value):
    index = hash(value)
    hashTable [index] = value

def contain(value):
    index = hash(value)
    return hashTable[index] == value

add("Bob")

print(contain("bob"))
print(hashTable )
```

```python
#When Collision happens use chaining
hashTable = [[],[],[],[],[],[],[],[],[],[]]
def hash(value):
    sum_of_char = 0
    for ch in value:
        sum_of_char += ord(ch)
    return sum_of_char % len(hashTable)

def add(value):
    index = hash(value)
    hashTable[index].append(value)

def contain(value):
    index = hash(value)
    return value in hashTable[index]


add('Bob')
add('Pete')
add('Jones')
add('Lisa')
add('Siri')
add('Stuart')
print(hashTable)
print(contain('Lisa'))
print(contain('Stuart'))
print(contain('Pete'))
```

```python
def check_isomorphic(str1, str2):
    dict1, dict2 = {},{}
    for ch1, ch2 in zip(str1, str2):
        if dict1.get(ch1, ch2) != ch2 and dict2.get(ch2, ch1) != ch1:
            return False
        dict1[ch1] = ch2
        dict2[ch2] = ch1
    return  True
print(check_isomorphic('ege', 'add'))
```

```python
# First non-repeating char in a string

def find_first_non_repeating_char(user_input):
    mydict = dict()
    for ch in user_input:
        mydict[ch] = mydict.get(ch, 0) +1
    for ch in user_input:
        if mydict[ch] == 1:
            return ch
    return None

user_input = input("Enter a string: ")
print(find_first_non_repeating_char(user_input))
```

```python
#sec largest number
def second_largest(lst):
    first = second = float('-inf')
    for num in lst:
       if num > first:
           second = first
           first = num
       elif num > second and num < first:
           second = num
    return second

print(second_largest([10,20,4,45,99]))
```

```python
#Compress a string => aaabbc=>a3b2c1
def compress_string(str1):
    str2 = ''
    mydict = dict()
    for ch in str1:
        mydict[ch] = mydict.get(ch, 0) +1
    for key, val in mydict.items():
        str2 += key+str(val)
    return str2

print(compress_string("aa!!bb@#"))

def compress_string_rle(str1):
    if not str1:
        return ""

    lst = []
    count = 1
    for i in range(1, len(str1)):
        if str1[i-1] == str1[i]:
            count += 1
        else :
            lst.append(str1[i-1])
            lst.append(str(count))
            count = 1
    lst.append(str1[-1])
    lst.append(str(count))
    return ''.join(lst)

print(compress_string_rle("aaabbca"))
```
