# Stack Implementation

### using Array

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self,item):
        self.stack.append(item)

    def isEmpty(self):
        return len(self.stack) == 0

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.stack.pop()

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.stack[-1]

    def size(self):
        return len(self.stack)

myStack = Stack()
myStack.push('Ali')
myStack.push('Yaqoob')
myStack.push(231)
myStack.size()
print('Poped element:',myStack.pop())


```

### Using LinkList

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None


class Stack:
    def __init__(self):
        self.head = None
        self.size = 0

    def push(self,val):
        new_node = Node(val)
        if self.head:
            new_node.next = self.head
            self.head = new_node
        else:
            self.head = new_node
        self.size += 1

    def traverseAndPrint(self):
        current = self.head
        while current:
            print(current.val,"->",end='')
            current = current.next

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        current = self.head
        self.head = self.head.next
        self.size -= 1
        return current.val

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.head.val

    def stack_size(self):
        return self.size

    def isEmpty(self):
        return self.size == 0


myNewObj = Stack()
myNewObj.push('Atif')
myNewObj.push('Adnan')
myNewObj.push('Usama')
myNewObj.push('Ali')
myNewObj.push('Usman')
myNewObj.push('Yaqoob')
myNewObj.traverseAndPrint()
print('\nPop value: ',myNewObj.pop())
print('Top value: ',myNewObj.peek())
print('isEmpty: ',myNewObj.isEmpty())
print('Top value: ',myNewObj.stack_size())
```
