# Stack Implementation

### using Array

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self,item):
        self.stack.append(item)

    def isEmpty(self):
        return len(self.stack) == 0

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.stack.pop()

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.stack[-1]

    def size(self):
        return len(self.stack)

myStack = Stack()
myStack.push('Ali')
myStack.push('Yaqoob')
myStack.push(231)
myStack.size()
print('Poped element:',myStack.pop())


```

### Using LinkList

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None


class Stack:
    def __init__(self):
        self.head = None
        self.size = 0

    def push(self, val):
        new_node = Node(val)
        if self.head:
            new_node.next = self.head
            self.head = new_node
        else:
            self.head = new_node
        self.size += 1

    def traverseAndPrint(self):
        if self.isEmpty():
            return "Stack is Empty"
        current = self.head
        while current:
            print(current.val,"->", end = '')
            current = current.next

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        current = self.head
        self.head = self.head.next
        self.size -= 1
        return current.val

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.head.val

    def stack_size(self):
        return self.size

    def isEmpty(self):
        return self.size == 0


myNewObj = Stack()
myNewObj.push('Atif')
myNewObj.push('Adnan')
myNewObj.push('Usama')
myNewObj.push('Ali')
myNewObj.push('Usman')
myNewObj.push('Yaqoob')
myNewObj.traverseAndPrint()
print('\nPop value: ',myNewObj.pop())
print('Top value: ',myNewObj.peek())
print('isEmpty: ',myNewObj.isEmpty())
print('Top value: ',myNewObj.stack_size())
```

## Queue Using LinkList

```python
class Node:
    def __int__(self, val):
        self.val = 0
        self.next = None


class Queue:
    def __init__(self):
        self.rear = None
        self.front = None
        self.size = 0

    def enqueue(self, val):
        newNode = Node(val)
        if self.rear == None:
            self.front = self.rear = newNode
        self.rear.next = newNode
        self.rear = newNode
        self.size += 1

    def isEmpty(self):
        return self.size == 0

    def dequeue(self):
        if self.isEmpty():
            return "Queue is Empty"
        temp = self.front
        self.front = self.front.next
        self.size -= 1
        return temp.val

    def peek(self):
        if self.isEmpty():
            return "Queue is Empty"
        return self.front.val

    def size(self):
        return self.size

    def traverse_and_print(self):
        if self.isEmpty():
            return "Queue is Empty"
        current = self.front
        while current:
            print(current.val,"->", end = '')
            current = current.next

myQueue = Queue()
myQueue.enqueue('Ali')
myQueue.enqueue('Yaqoob')
myQueue.enqueue(231)
myQueue.enqueue(34)
myQueue.enqueue(45)
myQueue.enqueue(56)
print(f'Poped values: {myQueue.dequeue()}->{myQueue.dequeue()}')
print(f'Peek Value: {myQueue.peek()}')
print("")
```

```python
#Hashing concept in Hash table to getting index
hash_set = ['Bob', "aqsa", 'Bob', 'Bob', 'Bob', 'Ali','Bob']
def hash_function(value):
    sum_of_char = 0
    for char in value:
        sum_of_char += ord(char)#convert characters into ASCII code
    return sum_of_char%10
print(f"index for BOB is: {hash_function('Aqsa')}")
def check_hash(value):
    index = hash_function(value)
    return  hash_set[index] == value
print(f'BOB contains: {check_hash("Bob")}')
```
